```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("SuperMarket Analysis.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns=df.columns.str.replace(" ","_")
df.columns=df.columns.str.lower()
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['date']=pd.to_datetime(df['date'])
df['time']=pd.to_datetime(df['time'],format="%I:%M:%S %p")
```

```python
df.info()
```

```python
#Basic Analysis
```

```python
#Total Sales
df["sales"].sum()
```

```python
#Average Sales
df["sales"].mean()
```

```python
#Total Transactions
total_trans=len(df)
total_trans
```

```python
#Average Rating
df["rating"].mean()
```

```python
#Customer Type Analysis using sales
df.groupby("customer_type")["sales"].sum().sort_values(ascending=False)
```

```python
#Profit By City
city_profit=df.groupby("city")["gross_income"].sum().sort_values(ascending=False)
city_profit
```

```python
#Categories Average Rating
rating_analysis=df.groupby("product_line")["rating"].mean().sort_values(ascending=False)
rating_analysis
```

```python
#Visualization
```

```python
#Sales by City
city_sales = df.groupby("city")["sales"].sum()
plt.figure(figsize=(5,5))
plt.pie(city_sales.values,labels=city_sales.index,autopct="%1.1f%%")
plt.title("Total Sales by City")
plt.figtext(0.5, -0.05, "Naypyitaw generated highest revenue among all cities(34.2%)", 
            wrap=True, horizontalalignment='center', fontsize=10, color='black')
plt.legend()
plt.show()
```

```python
#Sales by product_line
product_sales=df.groupby("product_line")["sales"].sum()
plt.figure(figsize=(15,6))
bars=plt.bar(product_sales.index,product_sales.values,color="green")
plt.title("Total Sales by Product line")
plt.xlabel("Product line")
plt.ylabel("Total Sales")
plt.bar_label(bars)
plt.figtext(0.5, -0.05, "Food and beverages is the top performing category", 
            wrap=True, horizontalalignment='center', fontsize=12, color='black')
plt.show()
```

```python
#Payment-Method Distribution
payment_type_sales=df["payment"].value_counts()
plt.figure(figsize=(5,5))
plt.pie(payment_type_sales.values,labels=payment_type_sales.index,autopct="%1.1f%%")
plt.title("Payment Method Distribution")
plt.figtext(0.5, -0.05, "E-wallet is the most preferred payment method(34.5%)", 
            wrap=True, horizontalalignment='center', fontsize=12, color='black')
plt.legend()
plt.show()
```

```python
#Month Wise Sales
df["month_name"] = df["date"].dt.month_name()
monthly_sales = df.groupby("month_name")["sales"].sum()
plt.figure(figsize=(5,5))
plt.pie(monthly_sales.values, labels=monthly_sales.index, autopct='%1.1f%%')
plt.title("Monthly Sales Distribution")
plt.figtext(0.5, -0.05, "January month has highest recorded sales(36.0%)", 
            wrap=True, horizontalalignment='center', fontsize=12, color='black')
plt.show()
```

```python
#Time Based Analysis
```

```python
df["hour"]=df["time"].dt.hour
hourly_sales=df.groupby("hour")["sales"].sum().sort_values(ascending=False)
plt.figure(figsize=(11,5))
bars=plt.bar(hourly_sales.index,hourly_sales.values,color="green")
plt.xlabel("Hours")
plt.ylabel("Sales")
plt.title("Sales by Hour")
plt.bar_label(bars,labels=[f'{(v/df["sales"].sum())*100:.1f}%' for v in hourly_sales.values])
plt.figtext(0.5, -0.05, "The most busy sale-hour is 7pm(12.3%)", 
            wrap=True, horizontalalignment='center', fontsize=12, color='black')
plt.show()
```

###  Final Insights Summary
*  **Top City:** Naypyitaw (34.2% Revenue & Profit)
*  **Payment:** E-wallet is most preferred (34.5%)
*  **Top Category:** Food & Beverages (Highest Sales & 7.11 Rating)
*  **Peak Month:** January (36.0% Sales)
*  **Peak Hour:** 7 PM is the busiest (12.3%)
